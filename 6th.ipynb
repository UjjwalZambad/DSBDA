```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"D:\downloads\archive (4)\forestfires.csv")
```

```python
data
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temp', y='area', data=data)
plt.xlabel('Temperature')
plt.ylabel('Burned Area')
plt.title('Temperature vs. Burned Area')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='month', y='area', data=data)
plt.xlabel('Month')
plt.ylabel('Burned Area')
plt.title('Monthly Burned Area')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.pairplot(data[['temp', 'wind', 'rain', 'month']], hue='month')
plt.suptitle('Pairplot of forest fire Dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['wind'], bins=20, kde=True)
plt.xlabel('Wind Speed')
plt.ylabel('Count')
plt.title('Distribution of Wind Speed')
plt.show()
```
