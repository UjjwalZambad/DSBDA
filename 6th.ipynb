```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"D:\downloads\archive (4)\forestfires.csv")
```

```python
data
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(data['temp'], data['area'], color='red', alpha=0.5)
plt.xlabel('Temperature (C)')
plt.ylabel('Burned Area')
plt.title('Temperature vs Burned Area')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='month', y='area', data=data)
plt.xlabel('Month')
plt.ylabel('Burned Area')
plt.title('Monthly Burned Area')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.pairplot(data[['temp', 'wind', 'rain', 'month']], hue='month')
plt.suptitle('Pairplot of forest fire Dataset')
plt.show()
```
