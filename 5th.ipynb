```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\Ujjwal\Desktop\heart.csv")
data
```

```python
data = data.replace('?', pd.NA).dropna().astype(float)
```

```python
sns.pairplot(data[['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'target']], hue='target', diag_kind='kde')
plt.suptitle('Pairplot of Heart Disease Dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Heart Disease Dataset')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data, x='age', hue='target', bins=20, kde=True)
plt.title('Distribution of Age by Heart Disease')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='target', data=data)
plt.title('Boxplot of Age by Sex and Heart Disease')
plt.show()
```
