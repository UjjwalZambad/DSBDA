```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\Ujjwal\Desktop\heart.csv")
data
```

```python
sns.pairplot(data[['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'target']], hue='target', diag_kind='kde')
plt.suptitle('Pairplot of Heart Disease Dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Heart Disease Dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='age', bins=20, kde=True, data=data)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='target', data=data)
plt.title('Boxplot of Age by Sex and Heart Disease')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='trestbps', data=data)
plt.title('Age vs. Resting Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Resting BP')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='target', data=data)
plt.title('Distribution of Target Variable')
plt.xlabel('Target')
plt.ylabel('Count')
plt.show()
```
